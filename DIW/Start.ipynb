```python
import GenSweepLine
import GenGcode
import math
```

```python
# 设置输出.csv文件名
geom_name='FoldFlower' 

# 绘制磁编程单元轮廓
path = [(0,0),(4,8*math.sin(60*(math.pi/180))),(4,18+8*math.sin(60*(math.pi/180))),(-4,18+8*math.sin(60*(math.pi/180))),(-4,8*math.sin(60*(math.pi/180)))]

# 设置扫描线间距，单位 mm
interval = 0.3

# 设置扫描角度，单位 °
angle_deg = 90

sweep1 = GenSweepLine.SweepLineGenerator(geom_name, path, interval, angle_deg)
sweep2 = sweep1.rotated(0,0,math.pi/3)
sweep3 = sweep2.rotated(0,0,math.pi/3)
sweep4 = sweep3.rotated(0,0,math.pi/3)
sweep5 = sweep4.rotated(0,0,math.pi/3)
sweep6 = sweep5.rotated(0,0,math.pi/3)

sweep1.addTable()
sweep2.addTable()
sweep3.addTable()
sweep4.addTable()
sweep5.addTable()
sweep6.addTable()

# sweep6.preview()
```

```python
# 设置G代码数据名称
data_name = "FoldFlower"

# 设定气压，单位 MPa
pressure = 0.35

# 打印速度，单位 mm
print_speed = 6

# 设置层高， 单位 mm
layerz = 0.35

# 设置层数
layern = 3

# 设置第一层层高，单位 mm
firstz = 0.1

# 设置断丝抬高，单位 mm
broken_wire_h = 2

# 设置提前出丝时间，单位 ms
advanced_on = 100

# 设置提前关丝距离，单位 mm
advanced_off = 0

generator = GenGcode.GcodeGenerator(data_name, pressure, print_speed, layerz, layern, firstz, broken_wire_h, advanced_on, advanced_off)
generator.write_gcode_file(r"E:\JupyterNotebook\DIW\Gcode\%s.txt" % data_name)
```
